```python
from executorlib_core.shell import ShellExecutor
from executorlib_core.subprocess import SubprocessExecutor
```

```python
with SubprocessExecutor(max_workers=1) as exe:
    future = exe.submit(["echo", "test"], universal_newlines=False, shell=False)
    print(future.result())
```

```python
with ShellExecutor(["python", "tests/executables/count.py"], universal_newlines=True) as exe:
    future = exe.submit(string_input="4", lines_to_read=5, stop_read_pattern=None)
    print(future.result())
```
